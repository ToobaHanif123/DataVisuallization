# Word Clouds:
It is also known as text clouds or tag clouds. It works in a simple way, the more a specific words appears in the source of textual data (such as speech, blogpost, database), the more words will emphasize in the word cloud. Word-clouds are used to perform high-level analysis and visualisation of text data.



### Importing package Word_cloud by **Andreas Mueller**:

```python
#import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS
```

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
from PIL import Image
```

## Let's play with these packages:

```python
#open and reading the alice's adventures in the wonderland novel
alice_novel = open('/content/alice_novel.txt', 'r').read()
```

```python
stopwords = set(STOPWORDS)
```

```python
#making the word cloud object
alice_wc = WordCloud(
    background_color = 'white',
    max_words = 3000,
    stopwords = stopwords
)
#generate the word cloud of alice novel
alice_wc.generate(alice_novel)
```

```python
alice_wc #word cloud is created and lets visualize it
```

```python
#display the wordcloud

plt.imshow(alice_wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

So, in the first 3000 words, the most common words are **Alice**, **said**, **time**, **Hatter**, **went** and so on.
as the **said** is not meaningful word so let's add it into the stopwords and regenerate the wordcloud.

```python
stopwords.add('said')

#regenerate the word cloud
alice_wc.generate(alice_novel)

#display the word cloud
fig = plt.figure(figsize = (14, 18))

plt.imshow(alice_wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

## Imposing Words onto the mask of any shape:
  Another cool thing you can implement with the `word_cloud` package is superimposing the words onto a mask of any shape. Let's use a mask of Alice and her rabbit. We already created the mask for you, so let's go ahead and download it and call it *alice_mask.png*.

```python
import numpy as np
##save the mask to alice mask
alice_mask = np.array(Image.open('/content/alice_mask.jpeg'))
```

```python
alice_mask
```

```python
##displaying the alice_mask
fig = plt.figure(figsize = (12, 16))

plt.imshow(alice_mask, cmap = plt.cm.gray, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

## Shaping the wordcloud according to the mask:

```python
#creating the alice_wordcloud object
alice_wcm = WordCloud(
    background_color = 'white',
    max_words = 3000,
    mask = alice_mask,
    stopwords = stopwords
)

#generte the wordcloud
alice_wcm.generate(alice_novel)
```

```python
#display the word cloud mask object

plt.imshow(alice_wcm, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```python
fig = plt.figure(figsize = (10, 14))

plt.imshow(alice_wcm, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

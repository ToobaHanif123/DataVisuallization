So each row consists of 13 features:
> 1. **IncidntNum**: Incident Number

> 3. **Descript**: Description of the crime or incident
> 4. **DayOfWeek**: The day of week on which the incident occurred
> 5. **Date**: The Date on which the incident occurred
> 6. **Time**: The time of day on which the incident occurred
> 7. **PdDistrict**: The police department district
> 8. **Resolution**: The resolution of the crime in terms whether the perpetrator was arrested or not
> 9. **Address**: The closest address to where the incident took place
> 10. **X**: The longitude value of the crime location
> 11. **Y**: The latitude value of the crime location
> 12. **Location**: A tuple of the latitude and the longitude values
> 13. **PdId**: The police department ID

```python
#get the first 100 crimes in df_incidents datframe
limit = 100

df_incidents1 = df_incidents.iloc[0:limit,:]

df_incidents1
```

```python
df_incidents1.shape
```

let's visualize from the map where these crimes take place by default francisco map style

```python
import folium
#San Franscisco latitutde and longitude

latitude = 37.77
longitude = -122.42

#create map instance

san_map = folium.Map(location = [latitude, longitude], zoom_start=12)

#displaying the map
san_map
```

now zoom the map and find the incident crime location. let's superimpose the locations of the crimes onto the map. The way to do that in **Folium** is to create a *feature group* with its own features and style and then add it to the `sanfran_map`.

```python
import folium
#San Franscisco latitutde and longitude

latitude = 37.77
longitude = -122.42

#create map instance

san_map = folium.Map(location = [latitude, longitude], zoom_start=12)

#Instantiate the feature group for the incidents in dataframe

incidents = folium.map.FeatureGroup()

#set the loop run through 100 crimes and then add each to the incidents feature group
for lat, long, in zip(df_incidents1.Y, df_incidents1.X):
  incidents.add_child(
  folium.features.CircleMarker(
          location = [lat, long],
          radius = 6,
          color = 'black',
          fill = True,
          fill_color = 'blue',
          fill_opacity = 0.6
      )
  )

#add incidents to map
san_map.add_child(incidents)
```

```python
import folium
#San Franscisco latitutde and longitude

latitude = 37.77
longitude = -122.42

#create map instance

san_map = folium.Map(location = [latitude, longitude], zoom_start=12)

#creating featurgroup instance
incidents = folium.map.FeatureGroup()

#set the loop run through 100 crimes and then add each to the incidents feature group
for lat, long, in zip(df_incidents1.Y, df_incidents1.X):
  incidents.add_child(
      folium.features.CircleMarker(
          location = [lat, long],
          radius = 6,
          color = 'red',
          fill = True,
          fill_color = 'yellow',
          fill_capacity = 0.6
      )
  )

#add pop up text to each marker on the map
latitudes = list(df_incidents1.Y)
longitudes = list(df_incidents1.X)
labels = list(df_incidents1.Category)

for lat, long, lab in zip(latitudes, longitudes, labels):
  folium.Marker([lat, long], popup = lab).add_to(san_map)

san_map.add_child(incidents)
```

this is really cool now it can be seen that what category of crime occurred at each marker!

####This also looking map so congested so to overcome this problem, there are two remedies:


### 1.Remove these circle markers and then just add the text to the circle markers themselves as follows:

```python
import folium
#San Franscisco latitutde and longitude

latitude = 37.77
longitude = -122.42

#create map instance

san_map = folium.Map(location = [latitude, longitude], zoom_start=12)

#creating featurgroup instance
incidents = folium.map.FeatureGroup()

#set the loop run through 100 crimes and then add each to the incidents feature group
for lat, long, label, in zip(df_incidents1.Y, df_incidents1.X, df_incidents1.Category):
  folium.features.CircleMarker(
      location = [lat, long],
      radius = 6,
      color = 'blue',
      fill = True,
      popup = label,
      fill_color = 'red',
      fill_capacity = 0.6
      ).add_to(san_map)


san_map
```

#### 2.Group the Markers into the different clusters. Each Cluster is then represented by the number of crimes in each neighbourhood.  

```python
from folium import plugins

import folium
#San Franscisco latitutde and longitude

latitude = 37.77
longitude = -122.42

#create map instance

san_map = folium.Map(location = [latitude, longitude], zoom_start=12)

#creating marker cluster instance
incidents = plugins.MarkerCluster().add_to(san_map)

#set the loop run through 100 crimes and then add each to the incidents feature group
for lat, long, label, in zip(df_incidents1.Y, df_incidents1.X, df_incidents1.Category):
  folium.features.CircleMarker(
      location = [lat, long],
      radius = 6,
      color = 'blue',
      fill = True,
      popup = label,
      fill_color = 'red',
      fill_capacity = 0.6
      ).add_to(incidents)

#display map
san_map
```

```python
import numpy as np 

Sales = [642.4, 721.5, 842.6, 895.3, 905.6, 951.2]
Years = [1991, 1992, 1993, 1994, 1995, 1996]

fig,ax = plt.subplots()
ax.plot(Years, Sales)
plt.title("Annual Sales of soft Rock")

ax.set_xlabel('YEARS')
ax.set_ylabel('SALES')

x_axis = np.arange(1991, 1996)
y_axis = np.linspace(642.4, 951.2, 10)

ax.set_xticks(x_axis)
ax.set_yticks(y_axis)

plt.show()
```

```python
Sales = [642.4, 721.5, 842.6, 895.3, 905.6, 951.2]
Years = [1991, 1992, 1993, 1994, 1995, 1996]
Purchase = [720.1, 821, 840.2, 900.7, 920.4, 954.3]

fig = plt.figure()
ax = fig.add_subplot()

ax.plot(Years, Sales)
ax.plot(Years , Purchase)
plt.title("Annual Sales of soft Rock")

ax.set_xlabel('YEARS')
ax.set_ylabel('SALES')

x_axis = np.arange(1991, 1997)
y_axis = np.linspace(642.4, 951.2, 10)

ax.set_xticks(x_axis)
ax.set_yticks(y_axis)

plt.show()
```

## Line Plots:

A line plot is a type of plot which displays information as a series of data points known as **makers** that connected with each other by straight line. It is used in continuous data. the best use of this line plot is in trend data set and in a series.

**Let's start with a case study:**

In 2010, Haiti suffered a catastrophic magnitude 7.0 earthquake. The quake caused widespread devastation and loss of life and about three million people were affected by this natural disaster. As part of Canada's humanitarian effort, the Government of Canada stepped up its effort in accepting refugees from Haiti. We can quickly visualize this effort using a `Line` plot:

**Question:** Plot a line graph of immigration from Haiti using `df.plot()`.

```python
haiti = df.loc['Haiti', years]
```

```python
haiti
```

```python
haiti.plot()
```

```python
haiti.plot()
```

```python
haiti.index = haiti.index.map(int) #changing the haiti index values from string to integer
```

```python
haiti.plot(kind = 'line')

plt.title("Immigration From Haiti")

plt.ylabel("Number of Immigrants")
plt.xlabel("Years")

plt.show()
```

Annotate the spike that can be clearly seen at 2010

```python
haiti.plot(kind = 'line')

plt.title("Immigration From Haiti")

plt.ylabel("Number of Immigrants")
plt.xlabel("Years")

#adding annotation
plt.text(2000, 6000, '2010 Earthqauke')

plt.show()
```

```python
CI = df.loc[['India', 'China'], years]
CI
```

```python
CI.plot(kind = 'line')
```

That doesn't look right...

Recall that *pandas* plots the indices on the x-axis and the columns as individual lines on the y-axis. Since `df_CI` is a dataframe with the `country` as the index and `years` as the columns, we must first transpose the dataframe using `transpose()` method to swap the row and columns.


```python
CI = CI.transpose()

CI
```

```python
CI.index = CI.index.map(int)
CI.plot(kind = 'line')

plt.title('Immigrants from China and India')
plt.ylabel("Number of Immigrants")
plt.xlabel("Years")

plt.show()
```

```python
Pak = df.loc['Pakistan', years]
Pak
```

```python
Pak.index = Pak.index.map(int) #mapping index to integer data type

Pak.plot(kind = 'line') #specifying the plot type

plt.title('Immigration From Pakistan to Canada')
plt.xlabel('Years')
plt.ylabel("Number of Immigrants")

#displaying the plot
plt.show()
```

```python
Pak.index = Pak.index.map(int) #mapping index to integer data type

Pak.plot(kind = 'bar') #specifying the plot type

plt.title('Immigration From Pakistan to Canada')
plt.xlabel('Years')
plt.ylabel("Number of Immigrants")

#displaying the plot
plt.show()
```

```python
Pak.index = Pak.index.map(int) #mapping index to integer data type

Pak.plot(kind = 'barh') #specifying the plot type

plt.title('Immigration From Pakistan to Canada')
plt.xlabel('Years')
plt.ylabel("Number of Immigrants")

#displaying the plot
plt.show()
```
